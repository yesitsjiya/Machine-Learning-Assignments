# Importing Library

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Importing Dataset

```python
dataset = pd.read_csv('mushrooms.csv')
```

# Plotting Bar Plot

```python
plt.figure(figsize=(10, 5))
sns.countplot(x='class', data=dataset)
plt.xlabel('Class')
plt.ylabel('Count')
plt.title('Class Distribution')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='class', hue='cap-shape', data=dataset , palette ='Pastel2')
plt.title('Distribution of Cap Shapes Across Different Types of Mushrooms')
plt.xlabel('Mushroom Class')
plt.ylabel('Count')
plt.legend(title='Cap Shape')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(x='class', hue = 'habitat', data=dataset, palette='spring')
plt.xlabel('Class')
plt.ylabel('Count')
plt.title('Class Distribution')
plt.show()
```

# Plotting ScatterPlot

```python
plt.figure(figsize=(20 , 10))

sns.scatterplot(x ='odor' , y = 'class' , data = dataset , color='palevioletred')
plt.title('Distribution of Odors Across Different Types of Mushrooms')

plt.title('Distribution of Odour based on types of Mushrooms')
plt.xlabel('Odor')
plt.ylabel('Class ')
plt.show()
```

```python
plt.figure(figsize=(20 , 10))

sns.scatterplot(x ='class' , y = 'stalk-shape' , data = dataset , color='crimson')
plt.title('Distribution of Stalk Shape Across Different Types of Mushrooms')


plt.xlabel('Class')
plt.ylabel('Stalk-shape ')
plt.show()
```

# Plotting Boxplot

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='cap-color', y='stalk-shape', data=dataset , color='lightgreen')  # Replace with relevant features
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='population', y='habitat', data=dataset , color='plum')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='stalk-surface-above-ring', y='stalk-surface-below-ring', data=dataset , color='mediumseagreen')
plt.xticks(rotation=45)
plt.show()
```

# Plotting Pie Chart

```python
plt.figure(figsize=(8, 6))

colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0', '#ffb3e6']

plt.pie(dataset['cap-color'].value_counts(),
        labels=dataset['cap-color'].value_counts().index,
        autopct='%1.1f%%',
        startangle=90,
        colors=colors)

plt.title('Distribution of Cap Color')
plt.show()
```

```python
plt.figure(figsize=(8, 6))

colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0', '#ffb3e6']


plt.pie(dataset['gill-spacing'].value_counts(),
        labels=dataset['gill-spacing'].value_counts().index,
        autopct='%1.1f%%',
        startangle=90,
        colors=colors)

plt.title('Distribution of Gill Spacing')
plt.show()
```

# Plotting Line Chart

```python
class_counts = dataset['class'].value_counts()

plt.figure(figsize=(8, 6))
plt.plot(class_counts.index, class_counts.values, marker='o', linestyle='-' , color='darkseagreen')
plt.xlabel('Mushroom Class')
plt.ylabel('Count')
plt.title('Count of Each Mushroom Class')
plt.grid(True)
plt.show()
```

```python
class_counts = dataset['ring-number'].value_counts()

plt.figure(figsize=(8, 6))
plt.plot(class_counts.index, class_counts.values, marker='o', linestyle='-' , color='palevioletred')
plt.xlabel('Mushroom ring number')
plt.ylabel('Count')
plt.title('Count of Each Mushroom ring number')
plt.grid(True)
plt.show()
```

```python
class_counts = dataset['ring-type'].value_counts()

plt.figure(figsize=(8, 6))
plt.plot(class_counts.index, class_counts.values, marker='o', linestyle='-' , color='teal')
plt.xlabel('Mushroom ring type')
plt.ylabel('Count')
plt.title('Count of Ring Type of Mushroom')
plt.grid(True)
plt.show()
```

```python
class_counts = dataset['gill-size'].value_counts()

plt.figure(figsize=(8, 6))
plt.plot(class_counts.index, class_counts.values, marker='o', linestyle='-' , color='slateblue')
plt.xlabel('Gill Size of Mushrooms')
plt.ylabel('Count')
plt.title('Count of Each Gill Size of Mushrooms')
plt.grid(True)
plt.show()
```
