```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('mall_customer.xlsx')
```

```python
df.head()
```

```python
df['Genre'] = df['Genre'].map({'Male': 0, 'Female': 1})
```

```python
df.head()
```

```python
plt.figure(figsize=(10, 4))
plt.scatter(df['Age'], df['Spending Score (1-100)'], c='green', label='Age vs Spending')
plt.xlabel('Age')
plt.ylabel('Spending Score')
plt.title('Scatter Plot: Age vs Spending Score')
plt.legend()
```

```python
plt.figure(figsize=(20,10))
sns.barplot(x=df['Annual Income (k$)'], y=df['Spending Score (1-100)'])
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score')
plt.title('Bar Plot')
```

```python
plt.figure(figsize=(5,8))
sns.boxplot(y=df['Age'], color='lightblue')
plt.title('Box Plot: Age Distribution')
```

```python

gender_counts = df['Genre'].value_counts()
plt.pie(gender_counts, labels=['Female', 'Male'], autopct='%1.1f%%', colors=['red', 'blue'])
plt.title('Pie Chart: Gender Distribution')
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(data=df, x='Age', y='Spending Score (1-100)', hue='Genre')
plt.xlabel('Age')
plt.ylabel('Spending Score')
plt.title('Line Chart: Age vs Spending Score')
```
