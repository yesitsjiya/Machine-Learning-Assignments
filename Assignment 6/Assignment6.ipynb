# Importing Library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Importing Dataset

```python
dataset = pd.read_csv('temperatures.csv')
```

# Data Preprocessing

Checking if their are null values in the dataset

```python
dataset.describe()
```

```python
dataset.info()
```

```python
dataset.isnull().sum()
```

```python
x = dataset['YEAR'].values.reshape(-1 , 1)
y = dataset['ANNUAL'].values.reshape(-1 , 1)
```

```python
print(x)
```

```python
print(y)
```

# Splitting the Dataset into traing set and test set

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.25 , random_state = 0)
```

# Printing training set and test set

```python
print(x_train)
```

```python
print(x_test)
```

```python
print(y_train)
```

```python
print(y_test)
```

# Training the Linear Regression model

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train , y_train)
```

# Predicting the results

```python
y_pred = regressor.predict(x_test)
print(np.concatenate((y_pred.reshape(len(y_pred) , 1) , y_test.reshape(len(y_test) , 1)) , 1))
```

# R-squared value

```python
from sklearn.metrics import r2_score
r2_score(y_test , y_pred)
```

# Mean Squared Error

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test , y_pred)
```

# Mean Absolute Error

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test , y_pred)
```

# Visualising Linear Model

```python
plt.scatter(x_train[:,0] , y_train , color = 'lightpink')
plt.plot(x_test[:,0] , y_pred , color='lightblue')
plt.title('Linear Regression')
plt.xlabel('Month')
plt.ylabel('Temperature')
plt.show()
```
